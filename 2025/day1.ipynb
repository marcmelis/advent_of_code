```python
with open('input1.txt') as file:
    lines = file.read().splitlines()
    instructions = [(line[0], int(line[1:])) for line in lines]
    
```

```python
current = 50
res = 0
for direction, distance in instructions:
    current += distance if direction == 'R' else -distance
    current = current % 100
    if current == 0:
        res+=1
res
```

```python
pos = 50
zero_passes = 0
for direction, distance in instructions:
    if direction == 'R':
        pos+=distance
        zero_passes+=pos//100
    elif direction == 'L':
        old_pos = pos
        pos-=distance
        if old_pos > 0 and pos <= 0:
            zero_passes += 1 + (-pos) // 100
        elif old_pos == 0:
            zero_passes += (-pos) // 100
    pos = pos%100
zero_passes
```
